```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
from pyspark.sql.functions import col

spark = SparkSession.builder \
    .appName("CompanyDepartmentAnalysis") \
    .getOrCreate()

# dataset (Company Employees)
data = [
    (1, "Arjun", "Finance", 40000, 2),
    (2, "Meera", "Technical", 60000, 1),
    (3, "Vikas", "HR", 30000, 3),
    (4, "Neha", "Technical", 70000, 2),
    (5, "Riya", "Finance", 45000, 1)
]

# schema
schema = StructType([
    StructField("Emp_ID", IntegerType(), True),
    StructField("Emp_Name", StringType(), True),
    StructField("Department", StringType(), True),
    StructField("Salary", IntegerType(), True),
    StructField("Working_Years", IntegerType(), True)
])


df = spark.createDataFrame(data, schema)
print("Employee Data:")
df.show()

# Create new column (Total Salary earned over years)
df_total = df.withColumn("Total_Earnings", col("Salary") * col("Working_Years"))

# Group by Department and calculate total earnings
result = df_total.groupBy("Department").sum("Total_Earnings")

# Display result
print("Department-wise Total Earnings:")
result.show()
spark.stop()
```
